```python
def reduce_mem_usage(df, fillna='min'):
    start_mem_usg = df.memory_usage().sum() / 1024**2 
    # print('Memory usage of properties dataframe is :',start_mem_usg,' MB')
    NAlist = [] # Keeps track of columns that have missing values filled in. 
    for col in df.columns:
        if df[col].dtype != object and df[col].dtype.type != np.datetime64 :  # Exclude strings
            
            # Print current column type
            # print('******************************')
            # print('Column: ',col)
            # print('dtype before: ',df[col].dtype)
            
            # make variables for Int, max and min
            IsInt = False
            mx = df[col].max()
            mn = df[col].min()
            
            # Integer does not support NA, therefore, NA needs to be filled
            if not np.isfinite(df[col]).all(): 
                NAlist.append(col)
            
            if fillna == 'min':
                df[col] = df[col].fillna(mn-1).clip(lower=0)
            elif fillna == 'ffill':
                df[col] = df[col].ffill().bfill()
                
            # test if column can be converted to an integer
            asint = df[col].fillna(0).astype(np.int64)
            
            # 我认为 sum 的方法比逐个比较要好,前面的代码保证了是整数
            result = (df[col] - asint)
            result = result.sum()
            if result > -0.01 and result < 0.01:
                IsInt = True
                
            # IsInt = np.all(np.isclose(df[col], asint, rtol=1e-05, atol=1e-08, equal_nan=True))
            
            try:
                # Make Integer/unsigned Integer datatypes
                if IsInt:
                    if mn >= 0:
                        if mx < 255:
                            df[col] = df[col].round().astype(np.uint8)
                        elif mx < 65535:
                            df[col] = df[col].round().astype(np.uint16)
                        elif mx < 4294967295:
                            df[col] = df[col].round().astype(np.uint32)
                        else:
                            df[col] = df[col].round().astype(np.uint64)
                    else:
                        if mn > np.iinfo(np.int8).min and mx < np.iinfo(np.int8).max:
                            df[col] = df[col].round().astype(np.int8)
                        elif mn > np.iinfo(np.int16).min and mx < np.iinfo(np.int16).max:
                            df[col] = df[col].round().astype(np.int16)
                        elif mn > np.iinfo(np.int32).min and mx < np.iinfo(np.int32).max:
                            df[col] = df[col].round().astype(np.int32)
                        elif mn > np.iinfo(np.int64).min and mx < np.iinfo(np.int64).max:
                            df[col] = df[col].round().astype(np.int64)
                
                # Make float datatypes 32 bit
                else:
                    df[col] = df[col].astype(np.float32)
            except Exception as e:
                print(col, f' ({mn}, {mx}) IsInt:{IsInt}, got error {e}')
            
            # Print new column type
            # print('dtype after: ',df[col].dtype)
            # print('******************************')
    
    # Print final result
    # print('___MEMORY USAGE AFTER COMPLETION:___')
    mem_usg = df.memory_usage().sum() / 1024**2
    print('Memory usage is: ',mem_usg,' MB')
    print('This is ',100*mem_usg/start_mem_usg,'% of the initial size')
    print(f'{NAlist} contain nan.')
    return df
```

```python


import pandas as pd
import numpy as np
import vectorbt as vbt
from mlquant.indicators.common import *


def flatten_columns(df):
    if isinstance(df.columns, pd.MultiIndex):
        columndict_lst = df.columns.to_flat_index().map(lambda x: dict(zip(df.columns.names,x)) )
        new_cols = ['&'.join(f'{k}:{v}' for k, v in d.items() ) for d in columndict_lst]
    else:
        new_cols = [f'{df.columns.name}:{i}' for i in df.columns]
    return new_cols

def flatten_rename_columns(df, prefix=None):
    columns = flatten_columns(df)
    if prefix:
        columns = ['{prefix}_{c}' for c in columns]
    df.columns = columns
    return df


stock = df.rename(columns={'vol':'volume'})
open = stock['open']
close = stock['close']
high = stock['high']
low = stock['low']
volume= stock['volume']
```
